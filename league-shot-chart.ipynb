```python
import matplotlib.pyplot as plt

from shot_chart import draw_court, get_league_shots, shot_chart
```

```python
plt.rc('font', family='serif')
```

```python
seasons = '2000-01', '2018-19'
shots = {s: get_league_shots(s) for s in seasons}
```

```python
season = seasons[1]
threshold = 100
shot_chart_options = {
    'threshold': threshold,
    'color': 'pct',
    'markersize': [25, 100],
    'nbins': 4,
    'gridsize': 50,
    'legend_args': {'columnspacing': 0.1},
}
title = f'NBA Leage-wide Shot Chart: {season}\nMinimum {threshold} Attempts'

fig, ax = plt.subplots(figsize=(12, 8), dpi=100)
shot_chart(shots[season], ax, title=title, **shot_chart_options)
draw_court(ax=ax)
plt.xlim(-260, 260)
plt.ylim(-60, 300)
plt.axis('off')
plt.show()
```

```python
threshold = 100

shot_chart_options = {
    'threshold': threshold,
    'color': 'pct',
    'markersize': [100, 400],
    'nbins': 4,
    'gridsize': 50,
    'legend_args': {'fontsize': 20, 'title_fontsize': 20},
    'title_args': {'fontsize': 30}
}
```

```python
def compare_shot_charts(shots, seasons, options):
    fig, axs = plt.subplots(ncols=len(seasons), figsize=(30, 12), constrained_layout=True)

    for season, ax in zip(seasons, axs):
        shot_chart(shots[season], ax, title=season, **options)
        draw_court(ax=ax)
        ax.set_xlim(-260, 260)
        ax.set_ylim(-60, 300)
        ax.axis('off')
        
    return fig, ax
```

```python
fig, ax = compare_shot_charts(shots, seasons, shot_chart_options)
fig.suptitle(f'NBA Leage-wide Shot Chart\nMinimum {threshold} Attempts', fontsize=40)
plt.show()
```

```python
shot_chart_options['color'] = 'efg'
fig, ax = compare_shot_charts(shots, seasons, shot_chart_options)
fig.suptitle(f'NBA Leage-wide Shot Chart\nMinimum {threshold} Attempts', fontsize=40)
plt.show()
```

```python
shot_chart_options['color'] = 'pts'
fig, ax = compare_shot_charts(shots, seasons, shot_chart_options)
fig.suptitle(f'NBA Leage-wide Shot Chart\nMinimum {threshold} Attempts', fontsize=40)
plt.show()
```

```python
topn = 200
shot_chart_options = {
    'topn': topn,
    'color': 'red',
    'markersize': 400,
    'gridsize': 50,
    'title_args': {'fontsize': 30}
}
fig, ax = compare_shot_charts(shots, seasons, shot_chart_options)
fig.suptitle(f'The Game Has Changed\nTop {topn} Shot Locations in the NBA', fontsize=40)
plt.show()
```
